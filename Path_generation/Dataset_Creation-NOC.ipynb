```python
def cleanprop(p):
    topic, sub = p.rsplit('.', 1)
    _, topic = topic.rsplit('.', 1)
    sub = sub.replace('_', ' ')
    topic = topic.replace('_', ' ')
    return topic, sub
```

```python
gold_train = json.loads(open('data/WebQSP.test.json').read())
prop_train = defaultdict(list)
count = 0
countc = 0
lists = []
for i in gold_train['Questions']:
    resetp = False
    for P in i['Parses']:
        if len(P['Constraints']) > 0:
            resetp = True
        print(P['Constraints'])
        ans = []
        props = []
        if P['InferentialChain']:
            props = P['InferentialChain']
        for a in P['Answers']:
            if a['AnswerType'] == "Value":
                ans.append(a['AnswerArgument'])
            else:
                ans.append((a['AnswerArgument'], a['EntityName']))
        print(count, i['RawQuestion'],'\n\t', P['TopicEntityName'], 
              P['TopicEntityMid'], '\n\t', 
              ans, '\n\t', 
              props)
        print()
    lists.append((i['RawQuestion'], int(resetp)))
    count +=1
    if resetp: countc +=1
    if count > 0:
        break
```

```python
gold_train = json.loads(open('data/WebQSP.test.json').read())
prop_train = defaultdict(list)
count = 0
countc = 0
lists = []

for i in gold_train['Questions']:
    count +=1
    resetp = False
    for P in i['Parses']:
        if len(P['Constraints']) > 0:
            resetp = True
        #print('Const', P['Constraints'])
        if True: ###### Constraint?
            ans = []
            if P['InferentialChain']:
                props = P['InferentialChain']
            if len(props) > 0: ###### property?
                #props = [make_chain(p) for p in props]
                for a in P['Answers']:
                    if a['AnswerType'] == "Value":
                        ans.append(a['AnswerArgument'])
                    else:
                        ans.append((a['AnswerArgument'], a['EntityName']))
                print(count, i['RawQuestion'],'\n\t', 
                      P['TopicEntityName'], P['TopicEntityMid'], '\n\t', 
                      ans, '\n\t',
                      props)
                lists.append((i['RawQuestion'], ' '.join(props)))
            print()
    if resetp: countc +=1
    if count > 1000000:
        break
```
